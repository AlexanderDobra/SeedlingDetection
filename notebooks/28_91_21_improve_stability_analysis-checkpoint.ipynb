## Validation MAP as test MAP approximation
It is also interesting to see how the validation MAP values align with the MAP values from the dataset (also for different validation sizes)

```python
groups = df2.groupby([("params", "valid_ratio")])
plot_num = 1
for group_val, group in groups:
    # We need to group all of the different values together here
    flat_valid = group["valid_MAP"].to_numpy().flatten()
    flat_test = group["test_MAP_step"].to_numpy().flatten()
    #Remove NaN values
    valid_nan = np.isnan(flat_valid)
    test_nan = np.isnan(flat_test)
    assert (valid_nan == test_nan).all(), "They should match"
    flat_valid = flat_valid[~valid_nan]
    flat_test = flat_test[~valid_nan]
    ax = plt.subplot(1, len(groups), plot_num)
    ax.title.set_text(f"validation_ratio {group_val}")
    ax.scatter(flat_valid, flat_test)
    ax.set_xlabel("validation MAP")
    ax.set_ylabel("test MAP")
    ax.axis([0, 1, 0, 1])
    ax.set_box_aspect(1)
    plot_num += 1
    # Add trendline
    z = np.polyfit(flat_valid, flat_test, 1)
    p = np.poly1d(z)
    xs = np.arange(0,1,0.01)
    plt.plot(xs,p(xs),"r--")
    plt.plot(xs,[p(1)]*len(xs), "b--")
    
plt.tight_layout()
plt.show()
```

Looks like the variation reduces with increasing validation ratio, as expected. However, the trend seems to be about 0.6. We can also look at individual runs

```python
groups = df2.groupby([("params", "valid_ratio")])
seed = 42
row_num = 1
NUM_SAMPLES = 4
basic_cols = ["test_MAP_step", "valid_MAP"]
for group_val, group in groups:
    # Take a sample of the runs of each group
    sample = group.sample(NUM_SAMPLES, random_state=seed)
    for i, index in enumerate(sample.index):
        transposed = multiindex_transpose(sample.loc[index:index, basic_cols])
        plot_ind = NUM_SAMPLES * (row_num - 1) + i + 1
        ax = plt.subplot(len(groups), NUM_SAMPLES, plot_ind)
        transposed.plot.line(ax=ax, legend=False)
        if ax.is_first_col():
            ax.set_ylabel(f"VR: {group_val}")
    row_num += 1
    
plt.tight_layout()
plt.show()
```
